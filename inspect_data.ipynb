Our project contains two part of inputs:


1.   one shot of target object with clear back ground
2.   target video sequence with the Apriltag as target anchor to render the target object into.

In order to train the segmentation model with proper data, we choose the Carvana dataset as our target object shot source.





```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('data/image.jpg')
plt.imshow(img)
plt.show()
```

```python
img = mpimg.imread('data/apriltag.png')
plt.imshow(img)
plt.show()
```

Sampled clips of the target video sequences are below. The sequence is with images with know camera matrix.

```python
fig=plt.figure(figsize=(8, 8))
columns = 2
rows = 5
for i in range(1, columns*rows +1):
    img = mpimg.imread('data/frames/frame00{}2.jpg'.format(i-1))
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```

We first try the Mask-RCNN for image segmentation. However, though in our application scenario the target image if relatively clean. The Mask-RCNN performed pretty bad.

Thus we replace it with a Unet for precision so that we can deal with the next steps in our pipeline.
